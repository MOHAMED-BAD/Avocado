```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('avocado.csv')
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0'] ,axis=1 , inplace=True)
```

```python
df.info()
```

```python
#Converting format of Date column
df['Date'] =pd.to_datetime(df.Date)
```

```python
#rename cols for better understand
df= df.rename(index=str, columns={"4046" : "Small Volume", "4225" : "Large Volume","4770" : "XLarge Volume" })
```

```python
df['Month'] = df.Date.dt.month_name()
```

```python
df['Day'] = df.Date.dt.day_name()
```

```python
#Null value
sns.heatmap(df.isnull(),yticklabels=False)
#No null value
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True , square=True)
#Show year in data and Year-D result from Date are the same
```

```python
df.head()
```

```python
#order cols
df = df[['region','type','Date','Day' , 'Month','year','Small Bags','Large Bags','XLarge Bags','Total Bags'
         ,'Small Volume' ,'Large Volume','XLarge Volume','Total Volume','AveragePrice']]
```

```python
df.head()
```

```python
df.hist(bins=30,figsize=(12,10),color='skyblue',ec="black")
```

```python
sns.relplot(x='year', y='AveragePrice', data=df ,kind="line")
#Show 2017 is the most year have AveragePrice
```

```python
sns.relplot(x='Month', y='AveragePrice', data=df ,kind="line" )
plt.xticks(rotation=90)
#Show October and September are the better month
#Show Februart is the Worst Month
```

```python
sns.relplot(x='Day', y='AveragePrice', data=df ,kind="line" )
#show all Day are Sunday
```

```python
df['Day'].value_counts()
```

```python
#Show Day Not add any information all Day are Sunday 
#Drop Day col
df.drop(['Day'],axis=1 , inplace =True)
```

```python
sns.relplot(x='Date', y='AveragePrice', data=df ,kind="scatter")
plt.xticks(rotation=90)
```

```python
df.head()
```

```python
#REL between Type
df['type'].value_counts().plot.pie(figsize=(6, 6), autopct='%.2f')
plt.title("REL between Type")
plt.ylabel('Number of type')
plt.xlabel('type') 
```

```python
sns.barplot(y="Total Volume", x="type", data=df, palette = 'Set3')
#Show Conventional Have large Total Volume
```

```python
sns.barplot(y="AveragePrice", x="type", data=df, palette = 'Set3')
#Show Organic is expensive and small
```

```python
sns.distplot(df['AveragePrice'],bins=10)
#most averageprice between 10-15
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x=df['region'], y=df['AveragePrice']);
plt.xticks(rotation=90)
#Show Sanfrancisco have the high averageprice
```

```python
sns.distplot(df['year'] ,kde=False)
#show 2018 are the less year for avocado
```

```python
region_sales = df.groupby('region')['Total Volume'].sum().drop(['TotalUS'])
regions_total = df['region'].unique()
#removing TotalUS entry
regions_total = np.delete(regions_total,-3)

plt.figure(figsize = (20,5))
plt.bar(regions_total,region_sales)
plt.xticks(rotation= 90)
plt.xlabel('Regions')
plt.ylabel('Total volume sold')
plt.title('Total volume sold region wise')
plt.show()
#Show California , SouthCenteral and West are most region sale Avocado
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=df["region"], y=df["AveragePrice"], palette="icefire_r")
plt.xticks(rotation=90)
plt.xlabel('Region')
plt.tick_params(labelsize = 15)
plt.ylabel('Average Price')
plt.title('Average Price of Avocado According to Region')
```
